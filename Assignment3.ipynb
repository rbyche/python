#Assignment 3-5

```python
def addall(arr):
    temp=0
    for i in range(len(arr)):
        temp+=arr[i]
    return temp
    
def addalllc(arr):
    return sum([i for i in arr])

def addallodd(arr):
    temp=0
    for i in range(len(arr)):
        if(arr[i]%2!=0):
            temp+=arr[i]
    return temp

def addalllcodd(arr):
    return sum([i for i in arr if i%2])

c=[1,2,3,4,5,6,7,8,9]
print "use for loop:", addall(c)
print "use list com:", addalllc(c)
print "use for loop:(odd)", addallodd(c)
print "use list com:(odd)", addalllcodd(c)
```

1. addall func: 리스트형을 매개변수로 받아 for문을 돌며 temp변수에 리스트의 값들의 누적값을 반환합니다.
2. addalllc func: 리스트형을 매개변수로 받아 리스트내포를 이용하여 리스트를 복사 후 sum합수를 통해 값의 합을 반환합니다.
3. addallodd func: addall func에 if문을 추가하여 temp변수에 홀수 리스트값들의 누적값을 반환합니다.
4. addalllcodd func: addalllc func에 if문을 추가하여 홀수값들만 리스트에 복사 후 sum합수를 통해 값의 합을 반환합니다.

#Incremental Project

```python
import urllib2
import string

sitelist=[]
sitelist.append("http://www.naver.com")
sitelist.append("http://www.daum.net")
sitelist.append("http://www.nytimes.com")
counterarr=[]

for list in sitelist:
    source = urllib2.urlopen(list).read()


    for i in range(0, source.count('<!--')):
        if source.find("<!--") != -1:
            source = source.replace(source[source.find("<!--"):source.find("-->")+3:],"")

    for i in range(0,source.count("<")):
        if source.find("<") != -1:
            source = source.replace(source[source.find("<"):source.find(">")+1:],"")

    words = source.split()

    counter={}
    i=0
    ilength=len(words)
    while i < ilength:
        j=0
        jlength=len(words[i])
        while j < jlength:
            if words[i][j] in string.punctuation:
                words[i]=words[i].replace(words[i][j], '')
                j-=1
                jlength=len(words[i])
            j+=1

        if len(words[i])==0:
            words.pop(i)
            ilength-=1
            i-=1
        else:
            if words[i] in counter:
                counter[words[i]]+=1
            else:
                counter[words[i]]=1
        i+=1 
    counterarr.append(counter)
    
for i in range(len(counterarr)):
    print
    print sitelist[i]*30
    print
    print counterarr[i].items()
```

words 리스트들의 원소들을 while문을 통해 각각 접근하여 for문을 이용하여 문자열 중 in string.punctuation에 true가 반환 할 경우 replace함수를 통해 제거합니다. 만약 제거 후 공백일 경우 pop하여 제거하며 counter사전에 해당 단어가 없으면 값을 1로 한 후 저장하고 존재할 경우 1씩 누적해 나아갑니다. 한 web site가 끝날 경우 리스트에 append한 후 다음 사이트로 이동합니다.
